```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['figure.figsize'] = [10, 10]
font = {'family' : 'consolas',
        'weight' : 'bold',
        'size'   : 18}
matplotlib.rc('font', **font)
```

```python
def plot_line(start,end,**kwargs):
    parameters = {"color"     : "black",
                  "linestyle" : "solid"}
    parameters.update(**kwargs)
    plt.plot((start[0],end[0]),(start[1],end[1]),**parameters)
    
def parametric_plot(func_x,func_y,start,end,numpoints=100,color="black"):
    parameter_axis = [start + (end-start)*float(i)/numpoints for i in range(numpoints) ]
    x_axis = [ func_x(ts) for ts in parameter_axis ]
    y_axis = [ func_y(ts) for ts in parameter_axis ]
    plt.plot(x_axis,y_axis,color=color)
```

```python
''' Spacetime diagram of Minkowsky spacetime '''

plot_line((-6,0),(6,0))
plot_line((0,-6),(0,6))

plot_line((-5.5,-5.5),(5.5,5.5),color="red",linestyle="dashed")
plot_line((-5.5,5.5),(5.5,-5.5),color="red",linestyle="dashed")

plt.annotate("$x$",(6,0.15))
plt.annotate("$t$",(0.15,6))

plt.annotate("IV",(4,0.25))
plt.annotate("I",(-0.5,4))
plt.annotate("II",(-4,-0.5))
plt.annotate("III",(-1,-4))



plt.show()
```

```python
''' Penrose Diagram of flat Minkowsky spacetime ''' 

scale = np.pi/2

plot_line((-scale,0),(scale,0))
plot_line((0,-scale),(0,scale))

plot_line((0,scale),(scale,0),color="red",linestyle="dashed")
plot_line((0,-scale),(scale,0),color="red",linestyle="dashed")
plot_line((0,-scale),(-scale,0),color="red",linestyle="dashed")
plot_line((0,scale),(-scale,0),color="red",linestyle="dashed")

plt.annotate("$x=\infty$",(scale-0.15,0.25))
plt.annotate("$x=\infty$",(-scale,0.35))
plt.annotate("$t=\infty$",(0.15,scale))
plt.annotate("$t=\infty$",(0.15,scale))
plt.annotate("$t=-\infty$",(0.15,-scale))

plt.annotate("$I^+$",(0.85,0.85))
plt.annotate("$I^+$",(-0.85,0.85))
plt.annotate("$I^-$",(-0.95,-0.85))
plt.annotate("$I^-$",(0.85,-0.85))


def plot_u_line(u):
    parametric_plot(lambda x: 0.5*(np.arctan(u+x)+np.arctan(u-x)),
                    lambda x: 0.5*(np.arctan(u+x)-np.arctan(u-x)),
                   -10,10)

def plot_v_line(v):
    parametric_plot(lambda x: 0.5*(np.arctan(x+v)+np.arctan(x-v)),
                    lambda x: 0.5*(np.arctan(x+v)-np.arctan(x-v)),
                   -10,10)

for u in range(-4,4):
    plot_u_line(u)
    plot_v_line(u)


plt.show()
```
