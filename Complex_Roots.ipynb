```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Canvas:
    def __init__(self,**kwargs):
        default = {"x_min":-1,
                   "x_max":1,
                   "y_min":-1,
                   "y_max":1,
                   "show_ticks":True}
        default.update(**kwargs)
        
        self._x_max = default["x_max"]
        self._x_min = default["x_min"]
        self._y_max = default["y_max"]
        self._y_min = default["y_min"]

        self._figures = []
    
    def show(self,**kwargs):
        default = {"density":1000}
        default.update(**kwargs)
        
        ''' Axes painting '''
        x_axis = [self._x_min + float(i/default["density"])*(self._x_max-self._x_min) for i in range(default["density"])]
        y_axis = [self._y_min + float(i/default["density"])*(self._y_max-self._y_min) for i in range(default["density"])]
        zeros  = [0 for _ in range(default["density"])]
        plt.plot(x_axis,zeros,color="black")
        plt.plot(zeros,y_axis,color="black")

        for figure in self._figures:
            figure.plot()

        plt.show()


    def add_figure(self,figure):
        self._figures.append(figure)
        
class Figure:
        bindings = {"scatter":plt.scatter,
                    "plot":plt.plot}
        def __init__(self,**kwargs):
            self.x=kwargs["x"]
            self.y=kwargs["y"]
            self.col = kwargs["col"]
            self.kind = kwargs["kind"]
        
        def plot(self):
            self.bindings[self.kind](self.x,self.y,color=self.col)
            



        
        

```

```python
C1 = Canvas(x_max=1)


n = 5/2
C1.add_figure(
    Figure(
    x = [np.cos(np.pi*i/n) for i in range(int(2*n+1))],
    y = [np.sin(np.pi*i/n) for i in range(int(2*n+1))],
    col = "red",
    kind = "plot")
)

C1.show()
```
